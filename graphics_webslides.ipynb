```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("mplStyleFS")

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
figs_path = f'/home/s3flsago/physik/MA/juggling-webslides/static/images'
```

```python
def three(hand='right', dots_per_beat=100):
    y_vals = np.linspace(0, 1, dots_per_beat*3)
    if hand=='left':
        y_vals = 1 - y_vals
    return y_vals

def one(hand='right', dots_per_beat=100):
    y_vals = np.linspace(0, 1, dots_per_beat*1)
    if hand=='left':
        y_vals = 1 - y_vals
    return y_vals
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,7)
dots_per_beat = 20
offset_throws = 2
n_throws = 15
hands = np.full(20, 'right')
hands[1::2] = 'left'

X = np.linspace(0,n_throws,dots_per_beat*n_throws)

y_ball1 = np.concatenate( [three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+1))])
ax.plot(X[2*dots_per_beat:dots_per_beat*(n_throws-0)], y_ball1[2*dots_per_beat:dots_per_beat*n_throws], c=colors[0], label='Ball 1', linewidth=5)

y_ball2 = np.concatenate( [1-three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+1))])
ax.plot(X[1*dots_per_beat:dots_per_beat*(n_throws-1)]+1, y_ball2[1*dots_per_beat:dots_per_beat*(n_throws-1)], c=colors[1], label='Ball 2', linewidth=5)

y_ball3 = np.concatenate( [three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+1))])
ax.plot(X[0*dots_per_beat:dots_per_beat*(n_throws-2)]+2, y_ball3[0*dots_per_beat:dots_per_beat*(n_throws-2)], c=colors[2], label='Ball 3', linewidth=5)

ax.set_ylim(-0.2, 1.2)
ax.set_yticks([0, 1], labels=['Right hand', 'Left hand'])
ax.set_xticks(np.arange(offset_throws,n_throws+1), labels=np.arange(n_throws+1-offset_throws))
ax.set_xlabel('t (throws)')
ax.legend()

fig.savefig(f'{figs_path}/3-diagram.png')
plt.show()
```

```python
fig2, ax2 = plt.subplots()
fig2.set_size_inches(20,7)

y_ball1 = np.concatenate( [three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+2))])
y_ball1[(offset_throws+4)*dots_per_beat:(offset_throws+8)*dots_per_beat] = 0
y_ball1[(offset_throws+8)*dots_per_beat:(offset_throws+n_throws-2)*dots_per_beat] = y_ball1[(offset_throws+8+2)*dots_per_beat:(offset_throws+8+7)*dots_per_beat]
ax2.plot(X[2*dots_per_beat:dots_per_beat*(n_throws-0)], y_ball1[2*dots_per_beat:dots_per_beat*n_throws], c=colors[0], label='Ball 1', linewidth=5)

y_ball2 = np.concatenate( [1-three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+2))])
y_ball2[(offset_throws+4)*dots_per_beat:(offset_throws+8)*dots_per_beat] = 1
y_ball2[(offset_throws+8)*dots_per_beat:(offset_throws+n_throws)*dots_per_beat] = y_ball2[(offset_throws+8+2)*dots_per_beat:(offset_throws+8+9)*dots_per_beat]
ax2.plot(X[1*dots_per_beat:dots_per_beat*(n_throws-1)]+1, y_ball2[1*dots_per_beat:dots_per_beat*(n_throws-1)], c=colors[1], label='Ball 2', linewidth=5)

y_ball3 = np.concatenate( [three(hand=hand, dots_per_beat=dots_per_beat) for hand, ind in zip(hands, range(int(n_throws/3)+2))])
y_ball3[(offset_throws+4)*dots_per_beat:(offset_throws+5)*dots_per_beat] = one(hand='right', dots_per_beat=dots_per_beat)
y_ball3[(offset_throws+5)*dots_per_beat:(offset_throws+n_throws-4)*dots_per_beat] = 1-y_ball3[(offset_throws+8+2)*dots_per_beat:(offset_throws+8+8)*dots_per_beat]
ax2.plot(X[0*dots_per_beat:dots_per_beat*(n_throws-2)]+2, y_ball3[0*dots_per_beat:dots_per_beat*(n_throws-2)], c=colors[2], label='Ball 3', linewidth=5)

ax2.set_ylim(-0.2, 1.2)
ax2.set_yticks([0, 1], labels=['Right hand', 'Left hand'])
ax2.set_xticks(np.arange(offset_throws,n_throws+1), labels=np.arange(n_throws+1-offset_throws))
ax2.set_xlabel('t (throws)')
ax2.legend()

fig2.savefig(f'{figs_path}/441-diagram.png')
plt.show()
```
